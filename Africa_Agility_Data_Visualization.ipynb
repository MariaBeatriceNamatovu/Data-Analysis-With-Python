### Histogram

```python
import numpy as np
# Generate random exam scores (normally distributed)
np.random.seed(42)  # for reproducibility
num_students = 100
exam_scores = np.random.normal(70, 10, num_students)  # mean=70, std=10

# Plot the histogram
plt.figure()
plt.hist(exam_scores, bins=20, edgecolor='black')
plt.show()
```

```python
exam_scores
```

`plt.hist(exam_scores, bins=20, color='skyblue', edgecolor='black')`: Plots a histogram of the exam scores. The bins parameter specifies the number of bins (intervals) to use in the histogram. Here, we use 20 bins.

Let's illustrate further using a DataFrame.
- Create a DataFrame containing the exam scores.
- Plot a histogram of the exam scores.

```python
# Generate random exam scores (normally distributed)
np.random.seed(42)  # for reproducibility
num_students = 100
exam_scores = np.random.normal(70, 10, num_students)  # mean=70, std=10

# Create DataFrame
df = pd.DataFrame({'Exam Score': exam_scores})

# Plot the histogram
plt.hist(df['Exam Score'],bins=20, color='skyblue', edgecolor='black')
plt.show()
```

### Box-and-Whisker Plots (Boxplots)

```python
# Generate random salaries for different departments
np.random.seed(42)  # for reproducibility

# Number of employees in each department
num_employees = 100
departments = ['HR', 'Finance', 'IT', 'Marketing']

# Generate random salaries for each department
salaries = []
for department in departments:
    salaries.extend(np.random.normal(loc=50000, scale=10000, size=num_employees))

# Create DataFrame
data = {'Department': np.repeat(departments, num_employees),
        'Salary': salaries}
df = pd.DataFrame(data)

# Create boxplot
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.boxplot(x='Department', y='Salary', data=df)
plt.xlabel('Department')
plt.ylabel('Salary')
plt.title('Distribution of Salaries Across Different Departments')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- Outliers: The boxplot identifies potential outliers in the data. Points outside the whiskers (the lines extending from the box) are considered outliers. These outliers may represent unusually high or low salaries within a department.

- Central Tendency: The boxplot allows us to observe the central tendency of salaries within each department. The central line within each box represents the median salary, which gives us an idea of the typical salary in each department.

- Variability: The length of the box (interquartile range, IQR) indicates the variability or spread of salaries within each department. A longer box implies greater variability in salaries.

- Skewness: The boxplot can also provide insights into the skewness of the salary distribution within each department. If one side of the box is longer than the other, it indicates skewness in that direction (positive skew if the longer side is on the right ((upper quartile)), negative skew if on the left ((lower quartile))).
