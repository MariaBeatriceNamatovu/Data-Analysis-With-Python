### Interaction Features:
We will explore creating interaction features by multiplying or dividing existing features.

Throughout the lab, we'll be starting with how to derive a new feature called the "stroke ratio," which is a measure of how efficiently an engine converts fuel into power.

```python
# Make a copy
autos = df.copy()
```

```python
# Create a new feature 'stroke_ratio'
autos['stroke_ratio'] = autos['stroke']/autos['bore']
```

```python
# Overview of the new feature and existing features
autos[['stroke','bore','stroke_ratio']].head()
```

We could create an `Horsepower * Engine Size` feature to capture the combined effect of these two variables.

```python
# Create the interaction feature
autos['horsepower_enginesize'] = autos['horsepower'] * autos['engine_size']
# Explore the new feature
autos['horsepower_enginesize']
```

The rationale behind creating this interaction feature is that the combined effect of horsepower and engine size might be more informative for predicting the target variable (e.g., price) than the individual features alone. The interaction feature can capture the synergistic relationship between these two variables.

### Aggregation
Group by Categorical Features:

Let's group the data by categorical features like `make` and calculate summary statistics such as the mean, median, or count of numerical features like `horsepower`, or `engine_size`.

```python
autos['make'].value_counts()
```

```python
# Group the data by 'make' and calculate summary statistics
model_stats = autos.groupby('make')[['horsepower', 'engine_size']].agg(['mean','median','count'])
# Overview
model_stats
```

```python
# Rename the columns for better readability
model_stats.columns = ['Horsepower_mean','Horsepower_median','Horsepower_count', 'Enginesize_mean','Enginesize_medain', 'Enginesize_count']

# Overview
model_stats
```

## 4. Feature Selection
After creating the new aggregated features, evaluate their importance and relevance to the target variable using techniques like correlation analysis or feature importance from tree-based models.

In our case we will just focus on the correlation analysis.

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation analysis works better on numeric variables

# Alternatively we can select the numeric variables we want to analyse
selected_numeric_columns = ['wheel_base',	'length',	'width',	'height',	'curb_weight',	'num_of_cylinders',	'engine_size',	'bore',	'stroke', 'horsepower','price']

autos_numeric = autos[selected_numeric_columns]

# Calculate the correlation matrix .corr()
corr_matrix = autos_numeric.corr()

corr_matrix
# Visualize the correlation matrix
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True,cmap ='coolwarm')
plt.show()
```
