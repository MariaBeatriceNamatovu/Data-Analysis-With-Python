```python
# Let's define a function to calculate z_score



x_scaled = []
def z_score(list):
  """
  Parameter :
            list: list of the number is want to scale

  Returns:
          list: scaled variables
  """
  mean = np.sum(list)/len(list)
  sd = np.std(list)
  for i in list:
    x_s = (i - mean)/sd
    x_scaled.append(x_s)

  return x_scaled
```

#### Standardization
- Standardization, also known as z-score normalization, transforms the data to have a mean of 0 and a standard deviation of 1.
- This ensures that features with different scales are comparable and can prevent certain features from dominating others during model training.

In general:
- Use `z-score` normalization when the data follows a normal distribution and you want to preserve the original distribution shape.

- We are utilizing the `StandardScaler` class from the `sklearn.preprocessing` module, which is a part of the `scikit-learn` library in Python.

```python
# Import the scikit-learn library

from sklearn.preprocessing import StandardScaler
```

The `scikit-learn` library, often abbreviated as `sklearn`, is a popular machine learning library in Python.

##### Summary
- `Z-score` is more robust to outliers compared to `min-max` scaling, as it uses the mean and standard deviation rather than the minimum and maximum values.
- `Min-max` scaling is more sensitive to outliers, as the minimum and maximum values can be heavily influenced by extreme data points.

### 5. Log Transformation
- Log transformation is used to stabilize variance, make the data more Gaussian-like, and improve the performance of models.
- It involves taking the logarithm of the values in a dataset, typically using the natural logarithm (base e) or the base 10 logarithm.
- You can implement log transformation using the NumPy library, which provides efficient mathematical functions for working with arrays.
- You can also implement log transformation using the pandas library in Python.

```python
# 1. Implementation in Numpy
import numpy as np

# Example dataset
sample_data = np.array([1,2,3,4,10,100,1000])

# Natural logarithm (base e) transformation
log_transformed = np.log(sample_data)

# Base 10 logarithm transformation
log10_transformed = np.log10(sample_data)

# print values
print(sample_data)
print(log10_transformed )
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data (exponential distribution)
np.random.seed(0)
sample_data = np.random.exponential(scale=1, size=1000)

# Apply log transformation
log_transformed_data = np.log(sample_data)

# Plot histograms for original and transformed data
plt.figure(figsize=(10, 6))

# Histogram of original data
plt.subplot(1, 2, 1)
plt.hist(sample_data, bins=30, color='blue', alpha=0.7)
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Histogram of log-transformed data
plt.subplot(1, 2, 2)
plt.hist(log_transformed_data, bins=30, color='orange', alpha=0.7)
plt.title('Log Transformed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# 2. Implementation in Pandas and Numpy

# Natural logarithm (base e) transformation for 'Fare' column
data['Age_log'] = data['Age'].apply(np.log)

# Base 10 logarithm transformation
data['Age_log10'] = data['Age'].apply(np.log10)
data['Age_log10']
```

- Above, we used the `apply()` function to apply the `np.log` function to each element in the data Series.
- The `apply()` function in pandas is used to apply a function along an axis of the DataFrame or Series. It allows one to apply a custom or built-in function to each element of the Series or DataFrame.
- `np.log` is a function from the NumPy library that calculates the natural logarithm (base e) of a given value or array-like object.
